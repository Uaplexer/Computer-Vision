```python
import cv2
import matplotlib.pyplot as plt
```

```python
def plot_rgb_histogram(img):
    for i, color in enumerate(('r', 'g', 'b')):
        plt.plot(cv2.calcHist(img, [i], None, [256], [0, 256]), color = color)
        plt.xlim([0, 256])
```

```python
img = cv2.imread('images/outside.JPG', 1)
```

```python
cv2.imshow('image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('images/outside2.jpg', img)
```

```python
plot_rgb_histogram(img)
```

```python
img_negated = 1 - img
```

```python
cv2.imshow('image_negated', img_negated)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
cv2.imwrite('images/outside_negated.jpg', img_negated)
```

```python
plot_rgb_histogram(img_negated)
```

The histograms of the original image show the natural distribution of pixel intensities across the red, green, and blue channels. Peaks in the histograms correspond to the most dominant brightness levels in the image, for example higher blue values for the sky or higher green values for vegetation.

When the image is negated, the histograms become mirrored with respect to the midpoint of the intensity axis (around 127.5). As a result, dark regions in the original image appear bright in the negative, and bright regions appear dark.

The overall shapes of the histograms remain the same, but they are reflected horizontally. For example, if the original histogram had a strong peak at lower intensities (darker tones), the negative histogram will show the same peak shifted to the higher intensity range (brighter tones). This demonstrates that negation preserves the frequency distribution of intensities while inverting their values.
